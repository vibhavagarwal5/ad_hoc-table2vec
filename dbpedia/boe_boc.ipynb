```python
def parallelize_dataframe(df, func, n_cores=10):
    df_split = np.array_split(df, n_cores)
    pool = multiprocessing.Pool(n_cores)
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df
```

```python
def late_fusion(table,query):
    s = []
    for i in query:
        for j in table:
            sim = cosine_similarity(i.reshape(1,-1),j.reshape(1,-1))
            s.append(sim)
    s = np.array(s).reshape(-1)
    return s
```

```python
def early_fusion(table,query):
    a = np.average(table,axis=0).reshape(1,-1)
    b = np.average(query,axis=0).reshape(1,-1)
    sim = cosine_similarity(a,b)
    return sim.reshape(-1)[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def compute(row):
    table_en_oh = np.array([l2[l1.index(i)] for i in row['table_entities_filter']])
    query_en_ih = row['query_entities_oh']
    return early_fusion(table_en_oh,query_en_ih), late_fusion(table_en_oh,query_en_ih)
```

```python
def get_categ(entities):
    l = []
    for i in entities:
        v = ent_w_cat[ent_w_cat['0_en'] == i]
        if v.shape[0]:
            l.append(v['2_cat_lst'].item())
        else:
            print(i)
    return l
```

```python
t = time.time()
def get_categ_p(d):
    d['table_categ'] = d.table_entities_filter.apply(get_categ)
    return d

z1 = parallelize_dataframe(z1, get_categ_p, 50)
print(time.time()-t)
```

```python
def generate_matix(categories):
    m = np.zeros((len(categories),len(all_categ)))
    for i in range(len(categories)):
        for j in categories[i]:
            try:
                m[i, all_categ.index(j)] = 1
            except:
                pass
    return m
```

```python
def compute_boc(row):
    table_cat_boc = generate_matix(row['table_categ'])
    query_cat_boc = generate_matix(row['query_categ'])
    return early_fusion(table_cat_boc,query_cat_boc), late_fusion(table_cat_boc,query_cat_boc)
```

```python
z1.iloc[:1].apply(lambda x: compute_boc(x),axis=1)
```

```python
t = time.time()
def compute_p(d):
    d['categ_fusion'] = d.apply(lambda x: compute_boc(x),axis=1)
    return d

z1 = parallelize_dataframe(z1, compute_p, 50)
print(time.time()-t)
```
